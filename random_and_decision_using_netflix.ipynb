```python
df = df.drop(columns=[
    'show_id',
    'cast'
])

df['duration'] = df['duration'].str.extract(r'(\d+)').fillna(0).astype(int)
```

```python
df.columns
```

```python
df['date_added'] = pd.to_datetime(df['date_added'].str.strip())
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
df.drop(columns='date_added', inplace=True)
```

```python
import pandas as pd
if pd.api.types.is_string_dtype(df['duration']):
    df['duration'] = df['duration'].str.extract(r'(\d+)').fillna(0).astype(int)
```

```python
df.head()
```

```python
df["content_age"] = 2026 - df["release_year"]
```

```python
df["duration_minutes"] = df["duration"]
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
col_1=['director','country','rating']
for col in col_1:
  df[col].fillna(df[col].mode()[0], inplace=True)
df.isnull().sum()
```

```python
df["year_added"].fillna(df["year_added"].median(), inplace=True)
df["month_added"].fillna(df["month_added"].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head(5)
```

```python
df["country"].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['type']=le.fit_transform(df['type'])
df['director']=le.fit_transform(df['director'])
df['country']=le.fit_transform(df['country'])
df['rating']=le.fit_transform(df['rating'])
df['listed_in']=le.fit_transform(df['listed_in'])
df['description']=le.fit_transform(df['description'])
df['title']=le.fit_transform(df['title'])
```

```python
df.columns
```

```python
df['rating'].unique()
```

```python
df['type'].unique()
```

```python
df.head()
```

```python
x=df.drop(['type','description','title','director','duration','duration_minutes'],axis=1)
y=df['type']
```

```python
x.columns
```

```python

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=2)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(accuracy_score(y_test,y_pred))
```

```python
print("test score",model.score(x_test,y_test))
print("train score",model.score(x_train,y_train))
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

**DECISION TREE**

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(max_depth=5)
dt.fit(x_train,y_train)
```
